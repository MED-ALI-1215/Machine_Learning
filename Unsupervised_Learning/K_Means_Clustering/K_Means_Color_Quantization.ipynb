```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpim
```

```python
im_arr = mpim.imread(r'C:\Users\lenovo\Desktop\Notebooks\DATA\palm_trees.jpg')
```

```python
plt.figure(dpi=150)
plt.imshow(im_arr)
```

```python
(h,w,c) = im_arr.shape
```

```python
h
```

```python
w
```

```python
c
```

```python
im_arr_2d = im_arr.reshape(h*w,c)
```

```python
im_arr_2d
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=6,random_state=101)
```

```python
labels = model.fit_predict(im_arr_2d)
```

```python
labels
```

```python
centers = model.cluster_centers_.round(0).astype(int)
```

```python
quantization = centers[labels]
```

```python
color_q = quantization.reshape(h,w,c)
```

```python
plt.imshow(color_q)
```
