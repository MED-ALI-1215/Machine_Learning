```python
from sklearn.cluster import KMeans
```

```python
from sklearn.cluster import DBSCAN
```

```python
def display (model,data):
    labels= model.fit_predict(data)
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels,palette='Set1')
```

```python
KM =KMeans(n_clusters=3) 
```

```python
display(KM,blobs)
```

```python
KM =KMeans(n_clusters=2) 
```

```python
display(KM,moons)
```

```python
display(KM,circles)
```

```python
DB = DBSCAN()
```

```python
display(DB,blobs)
```

```python
display(DB,moons)
```

```python
display(DB,circles)
```

```python
db = DBSCAN(eps=0.15)
```

```python
display(db,moons)
```

```python
display(db,circles)
```
